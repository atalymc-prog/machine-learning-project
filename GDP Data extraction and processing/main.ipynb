# URL den bilgi çekme

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.

```python
tables = pd.read_html(URL)

type(tables)
```

```python
df = tables[3]

type(df)

df.head()
```

```python
df.shape

df.columns
```

```python
df.columns = range(df.shape[1]) 
df.head()
```

```python
df = df[[0,2]]

df
```

```python
df = df.iloc[1:11,:]
df
```

```python
df.columns = ["Country", "GDP"]
```

```python
type(df["GDP"])
```

```python
df["GDP"]=df["GDP"].astype(int) 
```

Convert the GDP value in Million USD to Billions USD

```python
df[["GDP"]]=df[["GDP"]]/1000
```

```python
df
```

Use numpy.round() method to round the values to 2 decimal places.

```python
df[["GDP"]]=np.round(df[["GDP"]],2)
```

```python
df
```

```python
df.rename(columns={"GDP": "GDP (Billion USD)"})
```

Load the DataFrame to the CSV file named "Largest_economeis.csv"

```python
df.to_csv("./Largest_economies.csv")
```
