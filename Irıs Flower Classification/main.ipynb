# Modeling - Simple manual model
Manuel olarak dataya bakalım ve burada cutoff datayı ayırabileceğimiz cutoff değerini bulalım.

```python
def single_feature_prediction(petal_length):
    """ Predicts the species of iris flower based on petal length """
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

# Model Tuning

Model tuning, bir makine öğrenimi modelinin performansını optimize etmek için modelin hiperparametrelerini ayarlama sürecidir. Bu, modelin eğitim verilerine en uygun şekilde genelleme yapmasını ve yeni, görünmeyen veriler üzerinde daha iyi tahminler yapmasını sağlamak amacıyla yapılır.


```python
LogisticRegression?
```

```python
# Bu hücre, Lojistik Regresyon modelinin farklı C (regularizasyon parametresi) değerleri için çapraz doğrulama (cross-validation) kullanarak doğruluk oranlarını hesaplar.
# C parametresi, Lojistik Regresyon'da regularizasyon gücünün tersidir. Daha küçük C değerleri daha güçlü regularizasyon anlamına gelirken, daha büyük C değerleri daha zayıf regularizasyon anlamına gelir.
# Regularizasyon, modelin aşırı öğrenmesini (overfitting) engellemeye yardımcı olur.

for reg_param in (0.1, 0.5, 1, 2, 5, 10):
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv = 5, scoring="accuracy")
    print(f"C={reg_param}: Model accuracy: {np.mean(accuracies)*100:.2f}%")

    
```

### Bayes Hiperparametre Optimizasyonu

Bayes optimizasyonu, özellikle değerlendirilmesi maliyetli olan fonksiyonların (örneğin, büyük bir makine öğrenmesi modelinin eğitimi ve doğrulanması) global maksimumunu veya minimumunu bulmak için kullanılan bir stratejidir.

**Temel Çalışma Prensibi:**
1. **Amaç Fonksiyonu (Objective Function):** Optimize edilmek istenen, hiperparametrelere bağlı performans metriği (ör. doğrulama doğruluğu). Bu fonksiyonun gerçek yapısı bilinmez (black-box).
2. **Vekil Model (Surrogate Model):** Amaç fonksiyonunu yaklaşık olarak modelleyen olasılıksal bir modeldir (genellikle Gaussian Process kullanılır). Bu model, belirli bir hiperparametre setinin ne kadar iyi performans göstereceğine dair bir tahmin ve bu tahminin belirsizliğini (varyansını) sağlar.
3. **Edinim Fonksiyonu (Acquisition Function):** Bir sonraki adımda hangi hiperparametre setinin deneneceğine karar verir. Bu fonksiyon, "keşif" (exploration - belirsizliğin yüksek olduğu alanları deneme) ve "sömürü" (exploitation - modelin iyi sonuç vereceğini tahmin ettiği alanları deneme) arasında bir denge kurar.

**Adımlar:**
1. Başlangıçta rastgele seçilen birkaç hiperparametre noktası ile model eğitilir ve sonuçlar kaydedilir.
2. Mevcut veriler kullanılarak Vekil Model güncellenir.
3. Edinim Fonksiyonu maksimize edilerek bir sonraki en iyi aday hiperparametre seti seçilir.
4. Seçilen aday ile gerçek model eğitilir ve gerçek performans ölçülür.
5. Yeni veri noktası veri setine eklenir ve 2. adıma dönülür.

**Avantajı:** Grid Search veya Random Search'e göre daha az deneme ile daha iyi hiperparametre setlerine ulaşabilir, çünkü önceki denemelerden elde edilen bilgiyi (posterior) bir sonraki adımı yönlendirmek için kullanır.
